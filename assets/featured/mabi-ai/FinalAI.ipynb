## Original Nexon Dev AI
Default AI's
Default AIs can be switched to by either using the dialog window, or by typing special keywords into the chat box. The default AIs have a '*' symbol in front of their name. These cannot be directly modified, but players can create a duplicate of them and edit the copy.
These are the default AI's you can select:

Auto Attack Mode - Pet will target and battle enemies nearby on its own.
Collaborative Mode - Pet will attack with the master.
Healing Mode - Pet continuously uses Healing on its master.
Command Mode - Pet does nothing until commanded. See Pet Commands.

Note:
The three parts of an AI rule set
Event: Event defines why to activate a rule set.
Condition: Condition defines specific situations required for the rule set to be used.
Act: Act defines the actions the pet will perform when the rule set is activated.
Both event and condition must be true for the act to be processed.
Either the acts or the conditions can be omitted, but the event must always be defined.





## 1. Analysis: "Activate vs. Trigger"

The "Activate" (Prepare): The line <cmd name="prepare_skill" pet_skill="windmill" .../> only puts the pet into the spinning animation (loading the skill). It does not hit the enemy yet.

The "Wait" (The Danger Zone): The code forces a <cmd name="wait" min="2000" max="4000"/>. This is why the pet often looks like a "statue." It is waiting for the animation timer to finish.

The "Trigger" (Process): The actual damage only happens at <cmd name="process_skill" target="enemy" timeout="0"/>. If the enemy hits the pet during the 2–4 second "wait" period, the skill is canceled, and the pet never reaches the "trigger" command.

How this relates to our "Statue Bug"
In the original code, if the enemy stopped (preparing Counter), the pet would start this long Windmill sequence. However, if the enemy canceled their Counter and Ran at the pet while the pet was in that 2000ms to 4000ms wait period, the pet was helpless.


## 2. The misspelling is in the condition name skill_not_magicshiled_processing.

The engine actually expects shiled (with the e and i swapped) in many legacy versions. If you "correct" it to the proper English spelling (shield), the AI script will often fail to compile or the pet will ignore the condition entirely because the internal engine constant is hardcoded with that specific typo.

The Findings
Original Line: <condition name="skill_not_magicshiled_processing"/>

The Typo: shiled instead of shield.

The Verdict: Do not fix this. In our latest "Ship-Ready" XML, I preserved this exact spelling to ensure the engine recognizes the condition. If we changed it to the correct spelling, the pet would likely try to use Defense or Counter even while Magic Shield is active, leading to mana-drain and skill cancellation.
